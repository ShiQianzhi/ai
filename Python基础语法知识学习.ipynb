```python
# 核心数据结构：列表（list）、元组（tuple）、字典（dict）、集合（set）
```

```python
# 列表的使用
numbers = [1, 2, 3]
print(type(numbers), numbers)
# 防止下标越界
# 下标0第一个元素，-1:最后一个元素
numbers.append(4)
print(type(numbers), numbers)

# 移除最后一个元素
numbers.pop()
print(type(numbers), numbers)

# 移除某个元素（确保存在不然会报错）
numbers.remove(3) if 3 in numbers else None
print(type(numbers), numbers)
numbers.clear()
print(type(numbers), numbers)
```

```python
# 元组
numbers = (1, 1, 2, 3)
print(type(numbers), numbers)
# 列表和元组的区别就是，元组的元素值不支持修改，例如（numbers[0] *= 10）会报错
print(numbers.count(1))
print(type(numbers), numbers)

# 字典：又叫哈希表，是键值对的形式，元素唯一
numbers = {1: 'a', 2: 'b'}
print(type(numbers), numbers)

# 集合
numbers = {1, 2, 3}
print(type(numbers), numbers)
```
